```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from keras.layers import Dense, LSTM
```

```python
data=data.filter(['Close'])
dataset=data.values
dataset
#80% training, 20% test
training_data_len = int(np.ceil(len(dataset)*0.80))
training_data_len 
```

```python
#To prepare the historical stock price data for input into the LSTM model
#one step is scale historical stock price data to a range between 0 and 1
#second step is creating input and output: 
#The input value is the 60 most recent stock prices, and the output value is the next stock price after 60 days

#Scale our data from 0 to 1
scaler=MinMaxScaler(feature_range=(0,1))
scaled_data=scaler.fit_transform(dataset)
scaled_data
```

```python
#scaled data for training
train_data=scaled_data[0:training_data_len,:]
x_train=[]
y_train=[]
#train_data
for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i,0])
    #print(x_train)
    y_train.append(train_data[i,0])
    #print(y_train)
    if i<=61:
        print(x_train)
        print(y_train)
        
x_train, y_train = np.array(x_train), np.array(y_train)
#Afterwards, the x_train array is reshaped into a 3D array with dimensions (number of input sequences, 60, 1),
# where number of input sequences is the number of input sequences created in the for loop.
```

```python
model=Sequential()
model.add(LSTM(128,return_sequences=True,input_shape=(x_train.shape[1],1)))
```

```python
# Build LSTM model
model=Sequential()
model.add(LSTM(128,return_sequences=True,input_shape=(x_train.shape[1],1)))
model.add(Dropout(0.35))
# LSTM layers are followed by dropout layers that prevent overfitting and regularize the neural networks
model.add(LSTM(64, return_sequences=False))
model.add(Dropout(0.3))
model.add(Dense(25, activation='relu'))
model.add(Dense(1))
#Two LSTM layers, two dropout layers*, and two dense layers** make up the model architecture
#Complie the model
model.compile(optimizer='adam', loss='mean_squared_error',metrics=['accuracy'])
#Compiling a model in deep learning involves specifying the loss function, optimizer, and any metrics you want to track during training
model.fit(x_train,y_train,batch_size=1,epochs=21)
#batch_size is set to 1, meaning that the weights are updated after each input sequence, . It will run through the entire training data 21 times if the epochs parameter is set to 21
keras.utils.plot_model(model,'multi_input_and_output_model.png', show_shapes=True)
```
